Comparison between parental level of education and scores

```python
import scipy.stats as stats
```

```python
data_an = data
```

```python
data_an["parent_edu_level"] = data_an["parental level of education"].map({"high school":1,"some high school":2,"some college":3,
                                                                         "associate's degree":4,"bachelor's degree":5,"master's degree":6})
```

```python
data
```

Anova:
Null Hypothesis: There is no significant difference in marks among parental level of education.
Alternate Hypothesis: There is a significant difference in marks among parental level of education.
Confidence level=95%

```python
fvalue,pvalue=stats.f_oneway(data["parent_edu_level"],data["Total score"])
```

```python
print(pvalue)
```

Pvalues is less than 0.05 so we reject the null hypothesis by saying that there is a significant difference in 
marks among parental level of education
